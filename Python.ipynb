#Comentarios

Los comentarios nos sirven para poner anotaciones en nuestro código sin tener que añadirlas por separado. Nos sirven por ejemplo para incluir explicaciones de lo que estemos haciendo.

Hay dos tipos de comentarios:

Comentarios de una sola línea: se usa el símbolo # y no hace falta "cerrarlos". No pueden ocupar más de una línea.
Comentarios multilínea: se usa """ para abrir y """ para cerrar. Pueden ocupar una o varias líneas.

```python
x = 0
x = x + 1
print(x)
```

# Condicionales

A veces queremos hacer algo sólo cuando se cumple una condición. Para eso usamos la palabra reservada if seguida de una condición y acabado en dos puntos (:). Y en el caso contrario (si no se cumple la condición) usamos else:

```python

nombre = 'Ander'

if nombre in ('ander', 'Ander'):
  if nombre == 'Ander':
    print('Es con mayuscula.')
  else:
    print('Es con minuscula.')
  print('El nombre es ander o Ander')
else:
  print(f'El nombre es otro, es {nombre}.')
```

## While

Podemos usar una variable que vaya aumentando hasta un valor concreto. OJO: si no nos aseguramos de que en algún momento se va a dejar de cumplir la condición el bucle nunca parará (bucle infinito).

```python
for e in range(5, 15):
  n = e
  if len(str(e)) < 2:
    n = '0' + str(e)
  print(n)
```

# Diccionarios

Los diccionarios son conjuntos de elementos de tipo `clave: valor`.

Por ejemplo:

```python
mi_dic = {
    ',': 'C',
    '.': 'P',
    '(': 'D',
    ')': 'I'
}

un_string = 'Esto es una prueba que como veis tiene un String'
mi_lista = list(un_string)

pos = 0
for c in mi_lista:
  if c == ',':
    mi_lista[pos] = 'C'
  elif c == '.':    
    mi_lista[pos] = 'P'

  pos += 1

print(''.join(mi_lista))
```
